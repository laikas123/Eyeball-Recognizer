```python
base_path = "./69000/"


#origin of image is 0, 0 in the top left...
#keep that in mind
def draw_square_on_image(square_top_left_x, square_top_left_y, side_length, image_name):

    
    img = mpimg.imread(base_path + image_name + ".png")
    
    x_dimens = np.shape(img)[1]
    
    y_dimens = np.shape(img)[0]
    
    if(square_top_left_x + side_length >= x_dimens or square_top_left_x < 0 or square_top_left_x >= x_dimens):
        print("ERROR OUT OF X BOUNDS")
        return
    if(square_top_left_y + side_length >= y_dimens or square_top_left_y < 0 or square_top_left_y >= y_dimens):
        print("ERROR OUT OF Y BOUNDS")
        return
    
    
    #draw the top line
    x = [square_top_left_x, square_top_left_x+side_length]
    y = [square_top_left_y, square_top_left_y]
    
    plt.plot(x, y, color="white", linewidth=3)
    
    #draw the left line
    x = [square_top_left_x, square_top_left_x]
    y = [square_top_left_y, square_top_left_y + side_length]
    
    plt.plot(x, y, color="white", linewidth=3)
    
    #draw the right line
    x = [square_top_left_x+side_length, square_top_left_x+side_length]
    y = [square_top_left_y, square_top_left_y + side_length]
    
    plt.plot(x, y, color="white", linewidth=3)
    
    #draw the bottom line
    x = [square_top_left_x, square_top_left_x+side_length]
    y = [square_top_left_y+side_length, square_top_left_y+side_length]
    
    plt.plot(x, y, color="white", linewidth=3)
    
    
    imgplot = plt.imshow(img)
    plt.show()
    
    
    
draw_square_on_image(70, 50, 20, "69000")
```
